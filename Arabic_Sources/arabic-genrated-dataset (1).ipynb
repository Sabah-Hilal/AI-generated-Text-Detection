```python
import os
import json
import random

tashkeela_dir = "/kaggle/working/tashkeela_data/Tashkeela-arabic-diacritized-text-utf8-0.3"
sample_count = 5000
min_words=50
def extract_human_samples(dir_path, num_samples, min_words):
    human_samples = []
    for root, _, files in os.walk(dir_path):
        for file in files:
            if file.endswith(".txt"):
                with open(os.path.join(root, file), "r", encoding="utf-8", errors="ignore") as f:
                    content = f.read()
                    paragraphs = content.split('\n')
                    for para in paragraphs:
                        para = para.strip()
                        if len(para.split()) >= min_words:
                            human_samples.append({"text": para, "label": "human"})
                            if len(human_samples) >= num_samples:
                                return human_samples
    return human_samples

human_data = extract_human_samples(tashkeela_dir, sample_count,min_words)
len(human_data)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
import random
import os

model_name = "aubmindlab/aragpt2-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': '[PAD]'})
    model.resize_token_embeddings(len(tokenizer))

generation_config = {
    "max_length": 100,
    "num_return_sequences": 1,
    "do_sample": True,
    "top_k": 50,
    "top_p": 0.95,
    "temperature": 1.0,
    "pad_token_id": tokenizer.pad_token_id
}

prompts = [
    "في يوم من الأيام،",
    "ذات مساء هادئ،",
    "في بلدة صغيرة،",
    "في إحدى الليالي،",
    "كان هناك رجل حكيم،",
    "بين الجبال والوديان،",
    "في السوق القديم،",
    "في الصباح الباكر،"
]

os.makedirs("generated_ai_arabic", exist_ok=True)

output_file = "generated_ai_arabic/ai_arabic_5000.txt"
with open(output_file, "w", encoding="utf-8") as f:
    for i in range(5000):
        prompt = random.choice(prompts)
        input_ids = tokenizer.encode(prompt, return_tensors="pt")
        output_ids = model.generate(input_ids, **generation_config)
        text = tokenizer.decode(output_ids[0], skip_special_tokens=True)
        f.write(text.strip() + "\n")
        if i % 100 == 0:
            print(f"Generated {i} samples...")

print("✅ Done generating 5000 samples.")
```

```python
import pandas as pd

with open("/kaggle/working/human_arabic_5000.txt", "r", encoding="utf-8") as f:
    human_texts = [line.strip() for line in f if line.strip()]

with open("generated_ai_arabic/ai_arabic_5000.txt", "r", encoding="utf-8") as f:
    ai_texts = [line.strip() for line in f if line.strip()]

human_texts = human_texts[:5000]
ai_texts = ai_texts[:5000]

df = pd.DataFrame({
    "text": human_texts + ai_texts,
    "label": [0]*len(human_texts) + [1]*len(ai_texts)  # 0: Human, 1: AI
})

df = df.sample(frac=1).reset_index(drop=True)

output_path = "/kaggle/working/arabic_text_svc_dataset.csv"
df.to_csv(output_path, index=False, encoding="utf-8")

print(f"Combined dataset saved for SVC at: {output_path}")
```

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = set(stopwords.words('arabic'))

def clean_arabic_text(text):
    text = re.sub(r'[^ء-ي\s]', '', text)
    
    text = re.sub(r'[\u0649\u064A\u064A]', 'ي', text)  # Different forms of 'ya'
    text = re.sub(r'[\u0622\u0623\u0625\u0626]', 'ا', text)  # Different forms of 'alif'
    
    text = re.sub(r'\s+', ' ', text).strip()
    
    text = ' '.join([word for word in text.split() if word not in stop_words])
    
    return text

full_dataset = pd.read_csv("/kaggle/working/arabic_text_svc_dataset.csv")

def preprocess_with_pandas(dataset):
    dataset['text'] = dataset['text'].apply(clean_arabic_text)
    
    return dataset

cleaned_data = preprocess_with_pandas(full_dataset)

print(cleaned_data.head())

cleaned_data.to_csv("/kaggle/working/cleaned_arabic_text_svc_dataset.csv", index=False)
```
